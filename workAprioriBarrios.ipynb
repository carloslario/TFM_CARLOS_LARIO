```python
te = TransactionEncoder()
dt = pd.read_csv('databasev2_police.csv', header=None, dtype=str)
dataset = dt.values.tolist()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df, min_support=0.002, use_colnames=True)
rules = association_rules(frequent_itemsets, metric ="lift", min_threshold = 0.05) 
rules = rules.sort_values(['confidence'], ascending =[False])
```
