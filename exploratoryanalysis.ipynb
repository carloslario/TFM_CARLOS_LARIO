```python
import pandas as pd

# Función para obtener el resumen de cada columna
def resumen_columna(columna):
    # Calcular la frecuencia de cada valor único
    frecuencia_valores = columna.value_counts()
    # Calcular la proporción de cada valor único
    proporciones_valores = columna.value_counts(normalize=True)
    # Crear un DataFrame con los valores únicos, sus frecuencias y proporciones
    resumen = pd.DataFrame({'Valor': frecuencia_valores.index, 'Frecuencia': frecuencia_valores.values, 'Proporción': proporciones_valores.values})
    return resumen

# Obtener el resumen para cada columna en el dataset
for columna in dataset.columns:
    print(f"Resumen para la columna '{columna}':")
    print(resumen_columna(dataset[columna]))
    print()
```
