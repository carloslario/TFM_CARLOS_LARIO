```python
#import library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
#data set
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk',],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
te = TransactionEncoder()
```

```python
dt = pd.read_csv('pf_dataset.csv', header=None, dtype=str)
dataset = dt.values.tolist()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
frequent_itemsets = apriori(df, min_support=0.005, use_colnames=True)

frequent_itemsets
```

```python
frequent_itemsets['itemsets'] = frequent_itemsets['itemsets'].apply(lambda x: ', '.join(map(str, x)))
# Dividir los valores de los conjuntos de Ã­tems en columnas separadas
max_columns = frequent_itemsets['itemsets'].str.count(',').max() + 1
column_names = [f'Columna{i}' for i in range(1, max_columns + 1)]
frequent_itemsets[column_names] = frequent_itemsets['itemsets'].str.split(', ', expand=True)
frequent_itemsets.to_excel('aprioriresults.xlsx')
```

ECLAT

```python
#bulid association rules using support metric
rules = association_rules(frequent_itemsets, metric="support", support_only=True, min_threshold=0.005)

#use only support metric in Eclat algo using apriori
```

```python
rules
```

```python
rules = rules[['antecedents', 'consequents', 'support']]
rules
```

```python
# Convertir frozenset a lista y ordenar los Ã­tems, luego convertir a tupla
rules['antecedents'] = rules['antecedents'].apply(lambda x: tuple(sorted(list(x))))
rules['consequents'] = rules['consequents'].apply(lambda x: tuple(sorted(list(x))))

# Eliminar reglas duplicadas considerando los Ã­tems ordenados
rules = rules.drop_duplicates(subset=['antecedents', 'consequents'])

# Crear una funciÃ³n para separar Ã­tems en columnas
def split_items(rules, column):
    items = rules[column].apply(lambda x: list(x))
    max_len = items.apply(len).max()
    columns = [f'{column}_item_{i+1}' for i in range(max_len)]
    items_df = pd.DataFrame(items.tolist(), columns=columns)
    return items_df

# Separar los Ã­tems de 'antecedents' y 'consequents' en columnas diferentes
antecedents_df = split_items(rules, 'antecedents')
consequents_df = split_items(rules, 'consequents')

# Unir las columnas separadas con el DataFrame original
rules_expanded = pd.concat([antecedents_df, consequents_df, rules[['support']]], axis=1)

# Guardar el DataFrame expandido en un archivo Excel
rules_expanded.to_excel('association_rules_expanded.xlsx', index=False)

# Si deseas guardar los resultados en un nuevo archivo CSV tambiÃ©n
frequent_itemsets.to_csv('frequent_itemsets.csv', index=False)
rules.to_csv('association_rules.csv', index=False)
```
