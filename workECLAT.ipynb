```python
#librerias base
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
import itertools
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
from pyECLAT import ECLAT
```

```python
dataset = pd.read_csv('databasev2.csv', header=None).values.tolist()
dataset
```

```python
te = TransactionEncoder()
dt = pd.read_csv('databasev2.csv', header=None, dtype=str)
dataset = dt.values.tolist()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
```

```python
freq_intemsets= pd.DataFrame(dataset)
freq_intemsets
```

```python
eclat_class = ECLAT(data=freq_intemsets, verbose=True)
```

```python
eclat_class.df_bin
```

```python
get_ECLAT_indexes, get_ECLAT_supports = eclat_class.fit(min_support=0.01,
                                                           min_combination=2,
                                                           max_combination=4,
                                                           separator=' & ',
                                                           verbose=True)
```

```python
result = pd.DataFrame(get_ECLAT_supports.items(),columns=['Item', 'Support'])
result.sort_values(by=['Support'], ascending=False)
```

```python
result
```

```python
# Dividir la columna 'Item' por comas y expandir los resultados en nuevas columnas
result[['Item1', 'Item2', 'Item3', 'Item4']] = result['Itemset'].str.split(',', expand=True)

# Eliminar la columna original 'Item' si es necesario
result.drop(columns=['Itemset'], inplace=True)
result

```
